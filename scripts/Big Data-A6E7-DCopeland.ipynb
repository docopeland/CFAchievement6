```python
cont = []
region = []
for country in wdi['Country Name']:
    if country in africa:
        cont.append('Africa')
        if country in EAfrica:
            region.append('Eastern Africa')
        elif country in MAfrica:
            region.append('Middle Africa')
        elif country in NAfrica:
            region.append('Northern Arica ')
        elif country in SAfrica:
            region.append('Southern Africa')
        elif country in WAfrica:
            region.append('West Africa')
        else: region.append('check me')
    elif country in asia:
        cont.append('Asia')
        if country in WAsia:
            region.append('Western Asia')
        elif country in SAsia:
            region.append('Southern Asia')
        elif country in SEAsia:
            region.append('South-eastern Asia')
        elif country in EAsia:
            region.append('Eastern Asia')
        elif country in CAsia:
            region.append('Central Asia')
        else: region.append('check me')
    elif country in europe:
        cont.append('Europe')
        if country in NEurope:
            region.append('Northern Europe')
        elif country in WEurope:
            region.append('Western Europe')
        elif country in SEurope:
            region.append('Southern Europe')
        elif country in EEurope:
            region.append('Eastern Europe')
        else: region.append('check me')
    elif country in NAmerica:
        cont.append('North America')
        if country in caribbean:
            region.append('Caribbean')
        elif country in CAmerica:
            region.append('Central America')
        elif country in NNAmerica:
            region.append('Northern America')
        else: region.append('check me')
    elif country in oceania:
        cont.append('Oceania')
        if country in ANZ:
            region.append('Australia and New Zealand')
        elif country in melanesia:
            region.append('Melanesia')
        elif country in micronesia:
            region.append('Micronesia')
        elif country in polynesia:
            region.append('Polynesia')
    elif country in SAmerica:
        cont.append('South America')
        region.append('South America')
    else:
        cont.append('check me')
        region.append('check me')
```

```python
wdi['continent'] = cont
```

```python
wdi['region'] = region
```
