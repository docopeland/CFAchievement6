```python
#for loop to create a list to tag which indicators to keep and which to remove
keep = []
for val in wdi['Indicator Name']:
    if val in inIndicator:
        keep.append(True)
    else:
        keep.append(False)
```

```python
#adding the result to the df
wdi['Include'] = keep
```

```python
wdi['Include'].value_counts()
```

```python
#create a df that only has the wanted indicators
newWDI = wdi[wdi['Include'] == True]
```

```python
#grabbing the country names
countries = newWDI['Country Name'].unique().tolist()
```

```python
#setting the indext to country name and indicatoar name so that I can change the format
newWDI = newWDI.set_index(['Country Name','Indicator Name'])
```

```python
#drop country code, indicator code, unnamed: 66, and include columns
newWDI = newWDI.drop(columns = ['Country Code', 'Indicator Code','Unnamed: 66','Include'])
```

```python
#getting the data in the correct shape (indicators as variables)
newWDI = newWDI.stack().unstack(level = 1)
```

```python
#removing regions, e.g Arab World, Euro area
newWDI = newWDI['Afghanistan':]
```

```python
#resting the index so we can get country name back
newWDI = newWDI.reset_index()
```
